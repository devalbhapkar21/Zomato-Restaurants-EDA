```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('zomato.csv')
df.head(5)
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis = 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
def clean_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(clean_rate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

# Dropping the null values

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2People','listed_in(type)':'type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Both of these columns are almost same,thus dropping listed_in(city)

```python
df = df.drop(['listed_in(city)'],axis=1)
```

```python
df['Cost2People'].unique()
```

```python
def handle_commas(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['Cost2People'] = df['Cost2People'].apply(handle_commas)
df['Cost2People'].unique()
```

```python
df.head()
```

# Cleaning rest type columns

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts
```

# Cleaning locations column

```python
df.head()
```

```python
location = df['location'].value_counts(ascending = True)
location_lessthan300 = location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()    
```

```python
df.head()
```

# Cleaning cuisines column

```python
cuisines = df['cuisines'].value_counts(ascending=False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['type'].value_counts()
```
